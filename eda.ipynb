## Data Analysis

```python


bins = [0,5, 10, 15, 20, 25,30]
labels = ['0-5', '5-10', '10-15', '15-20', '20-25','25-30']

data['Projects Completed_Binned'] = pd.cut(data['Projects Completed'], bins=bins,labels=labels,right=False)
plt.figure(figsize=(10, 5))

sns.histplot(x='Projects Completed_Binned', data=data, hue='Productivity (%)')

plt.xlabel('Projects Completed')
plt.ylabel('')
plt.show()

```

### -->  When the number of projects completed reaches around 20-25, productivity improves drastically.

```python

bins = [0,20,40,60,80,100]
labels = ['0-20', '20-40', '40-60', '60-80','80-100']

data['Satisfaction_Binned'] = pd.cut(data['Satisfaction Rate (%)'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 5))
sns.countplot(x='Satisfaction_Binned',data=data,hue='Productivity (%)')

plt.xlabel('Satisfaction Rate (%) ')
plt.ylabel(' ')

plt.show()
```

### -->even with low satisfaction rate ,productivity is high

```python

bins = [20, 30, 40, 50, 60]
labels = ['20-30', '30-40', '40-50', '50-60']

data['Age_Binned'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 5))
sns.histplot(x='Age_Binned', data=data, hue='Productivity (%)')

plt.xlabel('Age')
plt.ylabel(' ')

plt.show()
```

### --> the age group of 30-40 is having higher productivity

```python
bins = [30000, 60000, 90000, 120000]

labels = ['30000-60000', '60000-90000', '90000-120000']


data['Salary_Binned'] = pd.cut(data['Salary'], bins=bins, labels=labels, right=False)


plt.figure(figsize=(10, 5))
sns.countplot(x='Salary_Binned', data=data, hue='Productivity (%)')
plt.xlabel('Salary')
plt.show()

```

### --> Productivity does not depend heavily on salary

```python
x = data.drop(['Age_Binned','Salary_Binned','Projects Completed_Binned'], axis=1)
x=x.drop(['Satisfaction_Binned'],axis=1)
x
```
